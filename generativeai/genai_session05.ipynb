```python
import re, time

# Start Cloudflare tunnel in the background
!cloudflared tunnel --url http://localhost:11434 --no-autoupdate > tunnel.log 2>&1 &

# Wait for the tunnel URL to appear in the log
url = None
for _ in range(30):  # wait up to ~30 seconds
    time.sleep(1)
    log = open("tunnel.log").read()
    match = re.search(r"https://[a-zA-Z0-9-]+\.trycloudflare\.com", log)
    if match:
        url = match.group(0)
        break

if url:
    print(" Cloudflare Tunnel is live at:\n", url)
else:
    print(" Tunnel URL not found. Check logs below:\n")
    print(log[-1000:])  # print last part of the log for debugging
```
